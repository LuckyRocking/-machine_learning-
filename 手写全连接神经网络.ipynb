```python
#搭建全连接神经网络
#批量梯度下降方法
import numpy as np
class SimpleDensenet():
    def __init__(self):
        self.weight1=np.random.random(1)
        self.weight2=np.random.random(1)
        self.weight3=np.random.random(1)
        self.weight4=np.random.random(1)
        self.weight5=np.random.random(1)
        self.weight6=np.random.random(1)
    #定义偏置项
        self.bias1=np.random.random(1)
        self.bias2=np.random.random(1)
        self.bias3=np.random.random(1)
    
    #定义激活函数
    def sigmoid(x):
        return 1/(1+np.exp(-x))
    
    #定义导数
    def divmod(x):
        return SimpleDensenet.sigmoid(x)*(1-SimpleDensenet.sigmoid(x))
    
    def forward(self,x):
        def sigmoid(x):
            return 1/(1+np.exp(-x))
        #前向传播
        x=np.array(x)
        h1=sigmoid(self.weight1*x[0]+self.weight3*x[1]+self.bias1)
        h2=sigmoid(self.weight2*x[0]+self.weight4*x[1]+self.bias2)
        y_pre=sigmoid(self.weight5*h1+self.weight6*h2+self.bias3)
        return h1,h2,y_pre
    
    def train(self,x,y_true,max_iter=100,learning_rate=0.1):
        x=np.array(x)
        for i in range(max_iter):
            h1,h2,y_pre=SimpleDensenet.forward(self,x)
            loss=(-1)*np.sum((y_true*np.log(y_pre))+(1-y_true)*np.log(1-y_pre))
            dw6=np.sum((-1)*((y_true/y_pre-(1-y_true)/(1-y_pre))*h1*y_pre*(1-y_pre)))
            dw5=np.sum((-1)*((y_true/y_pre-(1-y_true)/(1-y_pre))*h2*y_pre*(1-y_pre)))
            db3=np.sum((-1)*((y_true/y_pre-(1-y_true)/(1-y_pre))*y_pre*(1-y_pre)))
            dw4=np.sum(-1*((y_true/y_pre-(1-y_true)/(1-y_pre))*self.weight6*y_pre*(1-y_pre))*(h2*(1-h2))*x[1])
            dw1=np.sum(-1*((y_true/y_pre-(1-y_true)/(1-y_pre))*self.weight5*y_pre*(1-y_pre))*(h1*(1-h1))*x[0])
            dw2=np.sum(-1*((y_true/y_pre-(1-y_true)/(1-y_pre))*self.weight6*y_pre*(1-y_pre))*(h2*(1-h2))*x[0])
            dw3=np.sum(-1*((y_true/y_pre-(1-y_true)/(1-y_pre))*self.weight5*y_pre*(1-y_pre))*(h1*(1-h1))*x[1])
            db2=np.sum(-1*((y_true/y_pre-(1-y_true)/(1-y_pre)))*self.weight6*y_pre*(1-y_pre)*(h2*(1-h2)))
            db1=np.sum(-1*((y_true/y_pre-(1-y_true)/(1-y_pre)))*self.weight5*y_pre*(1-y_pre)*(h1*(1-h1)))
            self.weight1-=learning_rate*dw1
            self.weight2-=learning_rate*dw2
            self.weight3-=learning_rate*dw3
            self.weight4-=learning_rate*dw4
            self.weight5-=learning_rate*dw5
            self.weight6-=learning_rate*dw6
            self.bias1-=learning_rate*db1
            self.bias2-=learning_rate*db2
            self.bias3-=learning_rate*db3
            print(loss)
if __name__=='__main__':
    SimpleDensenet().train(x=np.array([[1,2,4],[3,4,5]]),y_true=np.array([0,1,1]))
```
